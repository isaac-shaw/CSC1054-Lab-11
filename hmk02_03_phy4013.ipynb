```python
"""
@author: isaacshaw
"""

""" imports and other global varibles """
import numpy as np
import matplotlib.pyplot as plt

P_0 = 101325 # pascals
T = 20 + 273.15 # room temp in C coverted to kelvin
M = 28.97 # g/mol molar mass of dry air
R = 8.31 # J / mol*K
g = 9.81 # m / s^2
```

```python
z = np.linspace(0,100)
P = P_0 * np.exp(-(M*g*z)/(R*T))

""" plot aesthetics """
fig, ax = plt.subplots()

ax.plot(z, P)

ax.set_ylabel("Pressure (Pa)")
ax.set_xlabel("vertical distance (m)")
ax.set_title("Atmospheric Pressure as a Function of Height")
```
